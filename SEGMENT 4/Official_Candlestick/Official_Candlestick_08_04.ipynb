```python
# Import dependencies
%matplotlib inline
import pandas as pd
import mplfinance as mpf
```

```python
# Import data
file = "AAPL_Pre_Official_DB_7_31.csv"
aapl_df = pd.read_csv(file, parse_dates=['Date'])
#aapl_df.set_index('Unnamed: 0',inplace=True)
#aapl_df.index.name=None
aapl_df.head()
```

```python
aapl_df = aapl_df.set_index("Date")
aapl_df.head()
```

```python
mpf.plot(aapl_df["1980"], volume=True)
```

```python
mpf.plot(aapl_df["2016-08":"2020-08"],figratio=(10,6),
         type="candle",
         title="Apple Share Price between 2016 and 2020",
         mav=(30),volume=True, 
         tight_layout=True,
         style="yahoo"
        )
```
